```python
import tkinter
from tkinter import *
from tkinter import ttk
from googletrans import Translator, LANGUAGES
import speech_recognition as sr

translator = Translator()
window = Tk()
window.title("Language Translator")
window.geometry("1000x600")
window.config(bg="#2929a3")

def lang1():
    global words
    x = Translator()
    result = x.translate(text=Input.get(1.0, END), src=src_lang.get(), dest=des_lang.get())
    output.delete(1.0, END)
    output.insert(END, result.text)

def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        audio = recognizer.listen(source)

    try:
        recognized_text = recognizer.recognize_google(audio)
        Input.delete(1.0, END)
        Input.insert(END, recognized_text)
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")

Label(window, text="Welcome to Language Translator App", height=3, width=50, font="arial 25 bold", bg="black",
      foreground="white").pack()
Label(window, text="Source Language", font="arial 15 bold", foreground="black").place(x=20, y=150)
Label(window, text="Output Language", font="arial 15 bold", foreground="black").place(x=750, y=150)
Input = Text(window, font="arial 15 bold", height=12, width=60)
Input.place(x=10, y=200)
output = Text(window, font="arial 15 bold", height=12, width=60)
output.place(x=750, y=200)
src_lan = list(LANGUAGES.values())
des_lan = list(LANGUAGES.values())
src_lang = ttk.Combobox(window, values=src_lan)
src_lang.place(x=200, y=150)
des_lang = ttk.Combobox(window, values=des_lan)
des_lang.place(x=950, y=150)
Button(window, text="Translate", font="arial 35 bold", command=lang1).place(x=600, y=550)
Button(window, text="Speech to Text", font="arial 20 bold", command=recognize_speech).place(x=400, y=550)
window.mainloop()
```
